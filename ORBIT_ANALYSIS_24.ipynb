```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# NSIGHT
nsight_control = pd.DataFrame(np.loadtxt('nsight_control.txt'))
nsight_exp3 = pd.DataFrame(np.loadtxt('nsight_exp3.txt'))
nsight_truth = pd.DataFrame(np.loadtxt('nsight_truth.txt'))
```

```python
len(np.arange(0,36.5,0.5))
```

```python
diff2 = []
diff3 = []
for i in np.arange(0,73,1):
    xyz2 = (nsight_truth[7][i]-nsight_control[7][i])**2 + (nsight_truth[8][i]-nsight_control[8][i])**2 + (nsight_truth[9][i]-nsight_control[9][i])**2
    xyz3 = (nsight_truth[7][i]-nsight_exp3[7][i])**2 + (nsight_truth[8][i]-nsight_exp3[8][i])**2 + (nsight_truth[9][i]-nsight_exp3[9][i])**2
    diff2.append(np.sqrt(xyz2))
    diff3.append(np.sqrt(xyz3))
    
plt.figure(figsize=(12,6))
plt.plot(np.arange(0,36.5,0.5), diff2, label='TIEGCM control')
plt.plot(np.arange(0,36.5,0.5), diff3, label='TIEGCM assimilated')
plt.legend()
plt.xlabel('Hours')
plt.ylabel('Position difference, $\Delta$r / km')
plt.title('Orbit propagation from UT01 on 23/06/2008 for an NSIGHT-1-like satellite')
plt.grid(True)
plt.show()
```
